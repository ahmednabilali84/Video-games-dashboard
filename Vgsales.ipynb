# Data Exploration

### Video Game Sales Dataset - Column Descriptions

| Column Name | Description |
|---|---|
| `Rank` | The ranking position of the video game based on global sales. Lower numbers indicate higher sales performance. |
| `Name` | The title or name of the video game. |
| `Platform` | The gaming platform or console on which the game was released, such as Wii, PS4, Xbox, or PC. |
| `Year` | The release year of the video game. Some rows may contain missing values if the release year is unknown. |
| `Genre` | The category or type of the game, such as Action, Sports, Racing, Role-Playing, or Shooter. |
| `Publisher` | The company responsible for publishing the video game. Some rows may contain missing values. |
| `NA_Sales` | Total sales of the game in North America, measured in millions of units. |
| `EU_Sales` | Total sales of the game in Europe, measured in millions of units. |
| `JP_Sales` | Total sales of the game in Japan, measured in millions of units. |
| `Other_Sales` | Total sales in other regions outside North America, Europe, and Japan, measured in millions of units. |
| `Global_Sales` | Total worldwide sales of the game, measured in millions of units. This is the sum of regional sales columns. |

### Dataset Summary

- **Number of Rows:** 16,598
- **Number of Columns:** 11
- **Dataset Topic:** Video game sales across different regions and platforms.

```python
def custom_detect_outlier(dataframe, col, lower_bound_percentage : float, upper_bound_percentage : float):

    'Return list of outliers indices of the given percentages'

    # Detect lower and upper bounds
    lower_bound = dataframe[col].quantile(lower_bound_percentage)
    upper_bound = dataframe[col].quantile(upper_bound_percentage)

    # Return outlier indices
    outliers_indices = dataframe[(dataframe[col] < lower_bound) | (dataframe[col] > upper_bound)].index
    return outliers_indicies
```

### Group by families

```python
def platform_family(p):
    if p in ['PS', 'PS2', 'PS3', 'PS4', 'PSP', 'PSV']:
        return 'PlayStation'
    elif p in ['X360', 'XB', 'XOne']:
        return 'Xbox'
    elif p in ['Wii', 'WiiU', 'GC', 'DS', '3DS', 'NES', 'SNES', 'N64', 'GB', 'GBA']:
        return 'Nintendo'
    elif p in ['PC']:
        return 'PC'
    else:
        return 'Other'

df['Platform_Family'] = df['Platform'].apply(platform_family)
```

```python
family_sales = df.groupby('Platform_Family')['Global_Sales'].sum().reset_index()
```

```python
px.bar(family_sales, x='Platform_Family', y='Global_Sales', title='Total Sales by Platform Family', text_auto= True).update_xaxes(categoryorder = 'max descending')
```

```python
px.pie(data_frame=df,names='Platform_Family',title='Genres by Total Global Sales',hole=0.35)
```
